```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
customer_df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
display(customer_df.shape)
customer_df.head()
```

```python
customer_df.isnull().sum()
```

```python
snake_case_mapping = {col: col.lower().replace(' ', '_') for col in customer_df.columns}
# Rename the columns using the dictionary
customer_df = customer_df.rename(columns=snake_case_mapping)
```

```python
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'], format='%m/%d/%y')
```

```python
customer_df.dtypes
```

```python
numerical_df = customer_df.select_dtypes(include='number')
numerical_df.nunique().sum()
```

```python
def differentiate_variables(df):
    continuous_cols = []
    discrete_cols = []

    for col in df.columns:
        if df[col].nunique() > 5000:  # Threshold for considering a variable continuous
            continuous_cols.append(col)
        else:
            discrete_cols.append(col)

    continuous_df = df[continuous_cols]
    discrete_df = df[discrete_cols]

    return continuous_df, discrete_df

# Assuming df is your DataFrame
continuous_df, discrete_df = differentiate_variables(numerical_df)

continuous_df.drop(columns=['income'], inplace=True)

discrete_df = pd.merge(discrete_df, numerical_df[['monthly_premium_auto', 'income']], on='monthly_premium_auto', how='left')
```

```python
# Calculate the correlation matrix
corr_matrix = numerical_df.corr()
# Set up the matplotlib figure
plt.figure(figsize=(10, 8))
# Plot the correlation matrix using a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
# Customize the plot
plt.title('Correlation Matrix')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
# Show plot
plt.show()
```

```python
#we can drop 2 variables with 0 correlation
```

```python

def plot_discrete_variables(df):
    # Loop through each column and plot
    for column in discrete_df.columns:
        plt.figure(figsize=(8, 6))
        sns.countplot(data=df, x=column)
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.xticks(rotation=45)
        plt.show()

def plot_continuous_variables(df):
    # Loop through each column and plot
    for column in continuous_df.columns:
        plt.figure(figsize=(8, 6))
        sns.histplot(data=df, x=column, kde=True)
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()

# Assuming df is your DataFrame containing both discrete and continuous variables
plot_discrete_variables(discrete_df)
plot_continuous_variables(continuous_df)
```

```python

# Define the condition for dropping values (e.g., values greater than a certain threshold)
condition = continuous_df['customer_lifetime_value'] > 70000 
condition2 = continuous_df['total_claim_amount'] > 2300
# Use boolean indexing to select rows where the condition is False (i.e., keep rows where the condition is not met)
continuous_df = continuous_df[~condition]
continuous_df = continuous_df[~condition2]
# 'filtered_df' now contains only the rows where the values in 'column' do not exceed the threshold
```
